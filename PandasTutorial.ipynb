Importing The Library

```python
import pandas as pd
import numpy as np
```

Creating a simple dataframe

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
print(data)
#A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns
#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)
```

scikit-learn comes with a few small standard datasets that do not require to download any file from some external website.


```python
# importing the boston house price data
from sklearn.datasets import fetch_california_housing
```

```python
#load boston_dataset in the variable
calf_dataset = fetch_california_housing()
print(type(calf_dataset))
```

Pandas Dataframe

```python
#pandas dataframe
calf_df = pd.DataFrame(calf_dataset.data, columns = calf_dataset.feature_names)
```
